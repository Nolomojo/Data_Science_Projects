```python
# import necessary libraries

import numpy as np
import pandas as pd

from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# function reads urls and extracts the HTML

def getHTMLContent(link):
    html = urlopen(link)
    soup = BeautifulSoup(html, 'html.parser')
    return soup
```

```python
# get list of countries by dependencies wikipedia url
# and parse through tables to find the table desired

content = getHTMLContent('https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population')
tables = content.find_all('table')
for table in tables:
    print(table.prettify())
```

```python
# After inspecting html 'wikitable sortable' 
# is the table we need

table = content.find('table', 
                     {'class': 'wikitable sortable'})
rows = table.find_all('tr')

# listing all links

for row in rows:
    cells = row.find_all('td')
    if len(cells) > 1:
        country_link = cells[1].find('a')
        print(country_link.get('href'))
```
